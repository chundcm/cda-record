## 柱状图/条形图/堆叠图(Bar)

柱状/条形图，通过柱形的高度/条形的宽度来表现数据的大小。

展示多个分类的数据变化和同类别各变量之间的比较情况。
* 适用：对比分类数据。
* 局限：分类过多则无法展示数据特点。
* 相似图表：
    1. 堆积柱状图。比较同类别各变量和不同类别变量总和差异。
    2. 百分比堆积柱状图。适合展示同类别的每个变量的比例。

```python
# 导入柱状图类
from pyecharts.charts import Bar
```

```python
attr = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
v1 = [23.0, 44.9, 57.0, 23.2, 25.6, 46.7, 35.6, 62.2, 32.6, 20.0, 36.4, 3.3]
```

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1)
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

### 调整柱间距离

```python
bar = Bar()
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1,category_gap = "50%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

### 改变颜色

```python
bar = Bar()
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1,category_gap = "50%",color = '#6495ED')
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

### 绘制多个类别

```python
attr = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
v1 = [23.0, 44.9, 57.0, 23.2, 25.6, 46.7, 35.6, 62.2, 32.6, 20.0, 36.4, 33.3]
v2 = [22.6, 45.9, 59.0, 26.4, 28.7, 50.7, 75.6, 82.2, 48.7, 18.8, 36.0, 2.3]
```

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%")
bar.add_yaxis('二号门店', v2,category_gap = "30%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

### 标记特殊点数据

```python
bar = Bar()
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%",
    markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_='max',name= '最大值')]) )
bar.add_yaxis('二号门店', v2,category_gap = "30%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

#### 添加多个标记点

```python
bar = Bar()
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%",
    markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_='max',name= '最大值'),
                                               opts.MarkPointItem(type_='min',name= '最小值'),
                                               opts.MarkPointItem(type_='average',name= '平均值'),]) )
bar.add_yaxis('二号门店', v2,category_gap = "30%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

#### 在多个类别中同时添加

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%")
bar.add_yaxis('二号门店', v2,category_gap = "30%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.set_series_opts(markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_='max',name= '最大值'),
                                               opts.MarkPointItem(type_='min',name= '最小值'),
                                               opts.MarkPointItem(type_='average',name= '平均值'),]) )

bar.render_notebook()
```

#### 改变标记点的形状和大小

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%")
bar.add_yaxis('二号门店', v2,category_gap = "30%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.set_series_opts(markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_='max',name= '最大值', symbol = 'arrow'),
                                               opts.MarkPointItem(type_='min',name= '最小值', symbol_size = 100),
                                               opts.MarkPointItem(type_='average',name= '平均值'),]) )

bar.render_notebook()
```

### 设置标记线

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%")
bar.add_yaxis('二号门店', v2,category_gap = "30%")
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.set_series_opts(markpoint_opts = opts.MarkPointOpts(data = [opts.MarkPointItem(type_='max',name= '最大值'),
                                               opts.MarkPointItem(type_='min',name= '最小值'),
                                               opts.MarkPointItem(type_='average',name= '平均值'),]) ,
                   markline_opts= opts.MarkLineOpts(data = [opts.MarkLineItem(type_ = 'average', name = '平均值'),
                                                           ]))

bar.render_notebook()
```

### 条形图

当我们类目名称比较长的时候，我们会做条形图(横向的柱状图)

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "50%")
bar.add_yaxis('二号门店', v2,category_gap = "50%")
# 翻转横竖两个坐标轴
bar.reversal_axis()
bar.set_series_opts(label_opts=opts.LabelOpts(position='right'))
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.render_notebook()
```

### 堆叠图

```python
bar = Bar()
```

```python
bar.add_xaxis(attr)
bar.add_yaxis('一号门店',v1 , category_gap = "30%", stack = '1')
bar.add_yaxis('二号门店', v2,category_gap = "30%" , stack = '1')
bar.set_global_opts(title_opts=opts.TitleOpts('2018年显示器销售额度(单位:百万)'))
bar.set_series_opts(label_opts= opts.LabelOpts(is_show = False))
bar.render_notebook()
```

### Pandas数据绘制一个柱状图

```python
import numpy as np
import pandas as pd
```

```python
drinks = pd.read_csv('data/drinks.csv', na_values='',keep_default_na= False)
```

```python
drinks
```

```python
# 统计每个大洲酒精消耗平均值
temp = drinks.groupby('continent').mean()['total_litres_of_pure_alcohol']
```

```python
temp
```

```python
temp.index
```

```python
list(temp.round(2))
```

```python
# 实例化图像
bar = Bar(init_opts= opts.InitOpts(width = '500px', height = '400px'))
```

```python
bar.add_xaxis(list(temp.index))
bar.add_yaxis('酒精消耗量' , list(temp.round(2)),category_gap='50%' )
bar.render_notebook()
```
