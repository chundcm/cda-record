# 常用图标绘制

## Gague（仪表盘）

仪表盘（guage，speedometer，dial chart，dashboard），是模仿汽车速度表的一种图表，常用来反映预算完成率、收入增长率等比率性指标。它简单、直观，人人会看，是商业面板（Dashboard）最主要特征的图表类型。

```python
from pyecharts import options as opts
```

```python
from pyecharts.charts import Gauge
```

```python
# 图片大小配置
size = opts.InitOpts(width='500px', height="500px")
```

```python
gauge = Gauge(size)
```

```python
gauge.add('课程录制图标',[('可视化课程录制完成率', '66.6')])
gauge.render_notebook()
```

### 绘制多个指标

```python
gauge = Gauge(size)
```

```python
gauge.add('课程录制图标',[('可视化课程录制完成率', '66.6'),('数据清洗',33.3)])
gauge.render_notebook()
```

## Funnel（漏斗图）

漏斗图在Google Analytics的报告里代表“目标和渠道”，在Web Trends里叫做“场景分析”，在Omniture的SiteCatalyst里被称为“产品转换漏斗”。虽然漏斗图称呼不一样，但它都是用来衡量网站中业务流程表现，并适用于电商等各个行业。漏斗图可以非常直观地看出网站业务流程中的问题所在，从而加以完善。

以电商为代表的网站（APP）：通过转化率比较能充分展示用户从进入网站到实现购买的最终转化率，漏斗图是评判产品健康程度的图表，由网站的每一个设计步骤的数据转化反馈得到结论，然后通过各阶段的转化分析去改善设计，提升用户体验的同时也提高网站的最终转化率。

```python
from pyecharts.charts import Funnel
```

```python
attr = ["访问", "注册", "购物车", "下单", "成功购买"]
value = [12000, 9000, 7000, 4000, 2000]
```

```python
# [("访问",12000),
# ('注册',9000)]  数据需要格式
```

```python
data = [(attr[i], value[i]) for i in range(len(attr))]
data
```

```python
funnel = Funnel(opts.InitOpts(width='600px', height="400px"))
```

```python
funnel.add('漏斗图',data, )
funnel.set_global_opts(title_opts=opts.TitleOpts('漏斗图测试'))
funnel.render_notebook()
```

### 更改标签位置

```python
funnel = Funnel(opts.InitOpts(width='600px', height="400px"))
```

```python
funnel.add('漏斗图',data, label_opts = opts.LabelOpts(position= 'left'))
funnel.set_global_opts(title_opts=opts.TitleOpts('漏斗图测试'))
funnel.render_notebook()
```

### 更改图例位置

```python
funnel = Funnel(opts.InitOpts(width='600px', height="400px"))
```

```python
funnel.add('漏斗图',data, )
funnel.set_global_opts(title_opts=opts.TitleOpts('漏斗图测试'),
                    legend_opts = opts.LegendOpts(orient = 'vertical', pos_left = 'left',pos_top = 'middle'))
funnel.render_notebook()
```

### 南丁格尔图

**南丁格尔玫瑰图**是弗罗伦斯·南丁格尔所发明的。又名**极区图**。是一种圆形的直方图。 南丁格尔自己常昵称这类图为**鸡冠花图（coxcomb）**，并且用以表达军医院季节性的死亡率，对象是那些不太能理解传统统计报表的公务人员。

**rosetype** -> str
是否展示成南丁格尔图，通过半径区分数据大小，有'radius'和'area'两种模式。默认为'radius'
* radius：扇区圆心角展现数据的百分比，半径展现数据的大小
* area：所有扇区圆心角相同，仅通过半径展现数据大小



```python
attr = ['Desktop','Imaging','Notebook','SNP','Server','Service']
v1 = [2291,1163,1616,513,3695,580]
v2 = [3311,1489,3509,163,2659,833]
```

```python
data4 = [(attr[i], v1[i]) for i in range(len(attr))]
data5 = [(attr[i], v2[i]) for i in range(len(attr))]
```

```python
data4
```

```python
pie = Pie(opts.InitOpts(width='500px',height = "500px"))
```

```python
pie.add('渠道A',data4, rosetype ="radius" )
pie.render_notebook()
```

```python
pie = Pie(opts.InitOpts(width='500px',height = "500px"))
```

```python
pie.add('渠道A',data4, rosetype ="area" )
pie.render_notebook()
```

### 把两个图绘制到一个图片上面

```python
pie = Pie(opts.InitOpts(width='800px',height = "500px"))
```

```python
pie.add('渠道A',data4, rosetype ="radius" ,center=["25%", "50%"], )
pie.add('渠道A',data5, rosetype ="area",center=["75%", "50%"],  )
pie.render_notebook()
```

```python
pie = Pie("饼图-玫瑰图示例", title_pos='center', width=1200)
pie.add("渠道A", attr, v1, center=[25, 50],  # 饼图的中心（圆心）坐标，数组的第一项是横坐标，第二项是纵坐标，默认为 [50, 50]
                                             # 默认设置成百分比，设置成百分比时第一项是相对于容器宽度，第二项是相对于容器高度
        is_random=True, radius=[30, 75],
        rosetype='radius', 
        is_label_show=True)
pie.add("渠道B", attr, v2, center=[70, 50], is_random=True, radius=[30, 75],
        rosetype='area',
        is_legend_show=False, is_label_show=True)
pie
```

### 嵌套环形图

```python
attr1 = ['直达', '营销广告', '搜索引擎']
attr2 = ['直达', '邮件', '联盟广告', '视频广告', '百度', '搜狗', '360','其他']
```

```python
v1 = [20,30,50]
v2 = [20,10,15,5,20,10,10,10]
```

```python
data6 = [(attr1[i], v1[i]) for i in range(len(attr1))]
data7 = [(attr2[i], v2[i]) for i in range(len(attr2))]
```

```python
data6
```

```python
data7
```

```python
pie = Pie(opts.InitOpts(width='500px',height = "500px"))
```

```python
pie.add('用户来源比例',data6, radius= ["50%", "60%"] )
pie.add('具体来源数据', data7 ,rosetype='radius' , radius=["0%",'40%'] ,
       label_opts = opts.LabelOpts(is_show = False))
pie.render_notebook()
```

### 两层环形图嵌套

```python
pie = Pie(opts.InitOpts(width='500px',height = "500px"))
```

```python
pie.add('用户来源比例',data6, radius= ["60%", "80%"] ,
        label_opts = opts.LabelOpts(position = 'inside',
                         formatter = "{b}: {d}%"))
pie.add('具体来源数据', data7, radius=["10%",'50%'] ,
       label_opts = opts.LabelOpts(position = 'inside',
                         formatter = "{b}: {d}%"))
pie.render_notebook()
```
