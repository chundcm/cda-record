## Geo：地理坐标系

省级地点和市级地点,县级地点,都能够支持

```python
from pyecharts.charts import Geo
```

```python
from pyecharts import options as opts
```

```python
from pyecharts.faker import Faker
```

```python
import pyecharts
```

```python
data = [('北京' , 250),
       ('上海',65),
       ('哈尔滨', 128),
       ('浙江', 25),
       ("杭州", 64)]
```

```python
geo = Geo()
```

```python
# 表计划
geo.add_schema(maptype='china') # 画中国地图
# 添加数据
geo.add('PM2.5', data)
geo.set_global_opts(title_opts= opts.TitleOpts(title = '全国各城市PM2.5统计值'))
geo.render_notebook()
```

## Geo地理图 手动新增坐标点

```python
data5 = [
        ("海淀区", 98),("朝阳", 120),("昌平区", 300),("天安门", 320)]
# ("四季青",22),
#         ("紫竹桥", 33),
#         ("车道沟",88),
```

```python
geo = Geo()
```

```python
# 表计划
geo.add_schema(maptype='北京' ) # 画中国地图
# 添加数据
geo.add('PM2.5', data5  ,type_='scatter')
geo.add_coordinate('天安门', 116.38 , 39.90)
geo.set_series_opts(label_opts= opts.LabelOpts(is_show = False))
geo.set_global_opts(title_opts= opts.TitleOpts(title = '全国各省PM2.5统计值'),
            visualmap_opts=opts.VisualMapOpts(),
            
        )
geo.render_notebook()
```
