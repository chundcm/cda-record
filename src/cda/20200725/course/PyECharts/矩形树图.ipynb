# TreeMap（树图）

树图是一种常见的表达『层级数据』『树状数据』的可视化形式。它主要用面积的方式，便于突出展现出『树』的各层级中重要的节点。

```python
import json
with open("treemap_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
# data
```

```python
from pyecharts import TreeMap

treemap = TreeMap("某品牌电脑产品线前三季度销量分布图", width=800, height=500)
treemap.add("Q1", data['data1'], is_label_show=True, treemap_left_depth=1, label_pos='inside')
treemap.add("Q2", data['data2'], is_label_show=True, treemap_left_depth=1, treemap_drilldown_icon='点击下钻', label_pos='outside')
treemap.add("Q3", data['data3'], is_label_show=True,
            treemap_visible_min=10000,
            label_pos='inside', legend_selectedmode='single')
treemap
```
