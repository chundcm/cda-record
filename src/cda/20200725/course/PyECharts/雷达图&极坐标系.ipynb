# Radar（雷达图）

雷达图主要用于表现多变量的数据。

集中划在一个圆形的图表上，来表现一个整体中的各项个体比率的情况,主要用于各项指标整体情况分析。（建议指标不要超过20项）

```python
from pyecharts import Radar

schema = [
    ("销售", 6500), ("管理", 16000), ("信息技术", 30000),
    ("客服", 38000), ("研发", 52000), ("市场", 25000)
]
v1 = [[4300, 10000, 28000, 35000, 50000, 19000]]
v2 = [[5000, 14000, 28000, 31000, 42000, 21000]]
radar = Radar()
radar.config(schema)
radar.add("预算分配", v1,
#           is_splitline=False,
#           is_axisline_show=True
         )
radar.add("实际开销", v2, label_color=["#4e97a7"], is_area_show=False,
          legend_selectedmode='single',
#           legend_selectedmode='multiple'
         )
radar
```

```python
from pyecharts import Radar

schema = [("推理力", 5), ("计算力", 5), ("观察力", 5), ("记忆力", 5), ("空间力", 5), ("创造力", 5)]
v1 = [[3.29, 3.59, 3.40, 2.78, 3.20, 2.93]]
v2 = [[2.66, 2.24, 2.10, 2.22, 3.28, 2.52]]
v3 = [[3.35, 3.52, 3.47, 3.77, 3.79, 3.37]]
radar = Radar()
radar.config(schema)
radar.add("李想", v1, is_splitline=True, is_axisline_show=True)
radar.add("薛飞", v2, label_color=["#4e79a7"], is_area_show=False)
radar.add("栾雨", v3, label_color=["#4e79d7"],
          is_area_show=False,
          legend_selectedmode='multiple'
         )
radar
```

# Polar（极坐标系）

```python
from pyecharts import Polar

radius = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
polar = Polar("极坐标系-堆叠柱状图示例")
polar.add("A", [1, 2, 3, 4, 3, 5, 1], radius_data=radius,
          type='barRadius', is_stack=True)
polar.add("B", [2, 4, 6, 1, 2, 3, 1], radius_data=radius,
          type='barRadius', is_stack=True)
polar.add("C", [1, 2, 3, 4, 1, 2, 5], radius_data=radius,
          type='barRadius', is_stack=True)
polar
```

```python
from pyecharts import Polar

radius = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
polar = Polar("极坐标系-堆叠柱状图示例", width=900, height=300)
polar.add("", [1, 2, 3, 4, 3, 5, 1], radius_data=radius,
          type='barAngle', is_stack=True)
polar.add("", [2, 4, 6, 1, 2, 3, 1], radius_data=radius,
          type='barAngle', is_stack=True)
polar.add("", [1, 2, 3, 4, 1, 2, 5], radius_data=radius,
          type='barAngle', is_stack=True)
polar
```
