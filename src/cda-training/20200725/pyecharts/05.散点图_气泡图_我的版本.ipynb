# Scatter（散点图）

直角坐标系上的散点图可以用来展现数据的 x，y 之间的关系，如果数据项有多个维度，可以用不同颜色来表现。

散点图表示因变量随自变量而变化的大致趋势，据此可以选择合适的函数对数据点进行拟合。顾名思义，就是散落的点去表达信息。
主要用于查找变量之间的相关性。

```python
import numpy as np
import pandas as pd
```

```python
from pyecharts.charts import Scatter
from pyecharts import options as opts
```

```python
# 设置图片大小变量
size= opts.InitOpts(width='500px',height='500px')
```

```python
x = ['衬衫', '毛衣', '领带', '裤子', '风衣', '高跟鞋', '袜子']
y = [61, 79, 105, 28, 39, 87, 33]
```

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis(x)
scatter.add_yaxis('销售额',y)
scatter.render_notebook()
```

### 默认x轴是分类型,如果直接传数值会出错

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis([1,2,3])
scatter.add_yaxis('测试',[1,2,3])
scatter.render_notebook()
```

### 把x轴设置为value类型

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis([1,2,3])
scatter.add_yaxis('测试',[1,2,3])
scatter.set_global_opts(xaxis_opts= opts.AxisOpts(type_='value'))
scatter.render_notebook()
```

### 生成数据绘制散点图

这里面有一个问题, 如果直接把这个整数数据传入进去,是不好使的

```python
v1 = np.random.randint(1,100,50)
```

```python
v2 = np.random.randint(1,100,50)
```

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis(v1)
scatter.add_yaxis('测试',v2)
scatter.render_notebook()
```

#### 转换成列表也不好使

```python
scatter = Scatter(size)
```

```python
# 不好使
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(list(v1))
scatter.add_yaxis('测试',list(v2))
scatter.render_notebook()
```

#### 如果用列表生成式生成数据

```python
import random
```

```python
x = [random.randint(1,100) for i in range(100)]
y = [random.randint(1,100) for i in range(100)]
```

```python
scatter = Scatter()
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(x)
scatter.add_yaxis('测试',y)
scatter.render_notebook()
```

#### 进一步测试,发现把数据类型改成浮点型就可以了

```python
v1 = np.random.randint(1,100,100)
v2 = np.random.randint(1,100,100)
```

```python
scatter = Scatter()
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(v1.astype('float'))
scatter.add_yaxis('测试' , v2.astype('float'))
scatter.render_notebook()
```

### 多类别数据散点图

```python
scatter = Scatter(size)
```

```python
# 如果只传一次x轴数据, 那么两组y轴数据将采用相同的x轴数据
scatter.add_xaxis([1,2,3])
scatter.add_yaxis('第一组实验对象',[1,2,3])
scatter.add_yaxis('第二组实验对象',[1.2,2.2,3.2])
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))
scatter.render_notebook()
```

```python
scatter = Scatter(size)
```

```python
# 如果要绘制两组不同的数据, 那么可以多次设置新的x轴数据和y轴数据
scatter.add_xaxis([1,2,3])
scatter.add_yaxis('男性',[1,2,3])
scatter.add_xaxis([1.2,2.2,3.2])
scatter.add_yaxis('女性',[1.2,2.2,3.2])
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))
scatter.render_notebook()
```

### 绘制房价和房屋面积直接的关系

```python
area = [30,40,50,60,70,80,90,100,120,150,200,300,400]
price = [1645,2745,3845,4945,5145,6450,7450,8450,9450,10450,13450,18000,24450]
```

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis(area)
scatter.add_yaxis('房价',price)
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))
scatter.render_notebook()
```

#### 去掉标签显示

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis(area)
scatter.add_yaxis('房价',price , label_opts = opts.LabelOpts(is_show = False))
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))
scatter.render_notebook()
```

#### 添加网格显示

```python
scatter = Scatter(size)
```

```python
scatter.add_xaxis(area)
scatter.add_yaxis('房价',price , label_opts = opts.LabelOpts(is_show = False))
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value',name = '面积',name_location ='center',name_gap = 30,
                           splitline_opts = opts.SplitLineOpts(is_show = True)),
                        yaxis_opts= opts.AxisOpts(name='房屋价格',name_location ='center',name_gap = 30 ,splitline_opts = opts.SplitLineOpts(is_show = True))
                       )
scatter.render_notebook()
```

### 添加视觉映射组件

利用 Visualmap 组件，通过颜色映射数值大小。

```python
np.random.seed(666)
v1 = np.random.randint(1,100,50)
v2 = np.random.randint(1,100,50)
```

```python
scatter = Scatter(size)
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(v1.astype('float'))
scatter.add_yaxis('测试' , v2.astype('float'))
scatter.set_global_opts(visualmap_opts = opts.VisualMapOpts())
scatter.render_notebook()
```

#### 气泡图: 将数据值映射为大小

```python
scatter = Scatter(size)
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(v1.astype('float'))
scatter.add_yaxis('测试' , v2.astype('float'))
scatter.set_global_opts(visualmap_opts = opts.VisualMapOpts(type_ = 'size'))
scatter.render_notebook()
```

#### 设置映射数据范围

```python
scatter = Scatter(size)
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(v1.astype('float'))
scatter.add_yaxis('测试' , v2.astype('float'))
scatter.set_global_opts(visualmap_opts = opts.VisualMapOpts(min_ = 30, max_=60))
scatter.render_notebook()
```

#### 设置为分段型映射

```python
scatter = Scatter(size)
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(v1.astype('float'))
scatter.add_yaxis('测试' , v2.astype('float'))
scatter.set_global_opts(visualmap_opts = opts.VisualMapOpts(is_piecewise = True))
scatter.render_notebook()
```

```python
scatter = Scatter()
```

```python
scatter.set_global_opts(xaxis_opts = opts.AxisOpts(type_='value'))

scatter.add_xaxis(v1.astype('float'))
scatter.add_yaxis('测试' , v2.astype('float'))
scatter.set_global_opts(visualmap_opts = opts.VisualMapOpts(is_piecewise = True,
                                            pos_left = '0%'))
scatter.render_notebook()
```

### EffectScatter（带有涟漪特效动画的散点图）
利用动画特效可以将某些想要突出的数据进行视觉突出。

```python
from pyecharts.charts import EffectScatter
```

```python
attr = ['毛衣','连衣裙','外套','皮裤','内衣','长裙']
```

```python
v1 = [10, 20, 30, 40, 50, 60]
v2 = [25, 40, 15, 10, 60, 33]
```

```python
effect = EffectScatter(size)
```

```python
effect.add_xaxis(attr)
effect.add_yaxis('测试',v2)
effect.render_notebook()
```

#### EffectScatter-不同Symbol

- ARROW 箭头
- DIAMOND 钻石
- RECT 方形
- ROUND_RECT 圆角方形
- TRIANGLE 三角形

```python
import pyecharts
```

```python
from pyecharts.globals import SymbolType
```

```python
pyecharts.globals.SymbolType
```

```python
# 可用的形状
dir(pyecharts.globals.SymbolType)[:5]
```

```python
effect = EffectScatter()
```

```python
effect.add_xaxis(attr)
effect.add_yaxis("测试", v2 , symbol=SymbolType.ARROW)
effect.render_notebook()
```

### 多组数据显示成不同的涟漪特效

```python
effect = EffectScatter()
effect.add_xaxis(attr)
effect.add_yaxis("测试", v1 , symbol=SymbolType.DIAMOND,symbol_size=20)

effect.add_yaxis("测试", v2 , symbol=SymbolType.TRIANGLE,symbol_size = 30 )

effect.render_notebook()
```
