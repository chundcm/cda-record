## 箱线图

箱形图是一种用作显示一组数据分散情况资料的统计图。它能显示出一组数据的最大值、最小值、中位数、下四分位数及上四分位数。

```python
from pyecharts.charts import Boxplot
```

```python
from pyecharts import options as opts
```

```python
size = opts.InitOpts(width="500px" ,height= '500px')
```

```python
x_axis = ['一月份', '二月份', '三月份', '四月份', '五月份']
y_axis = [
    [850, 740, 900, 1070, 930, 850, 950, 980, 980, 880,1000, 980, 930, 650, 760, 810, 1000, 1000, 960, 960],
    [960, 940, 960, 940, 880, 800, 850, 880, 900, 840,830, 790, 810, 880, 880, 830, 800, 790, 760, 800],
    [880, 880, 880, 860, 720, 720, 620, 860, 970, 950,880, 910, 850, 870, 840, 840, 850, 840, 840, 840],
    [890, 810, 810, 820, 800, 770, 760, 740, 750, 760,910, 920, 890, 860, 880, 720, 840, 850, 850, 780],
    [890, 840, 780, 810, 760, 810, 790, 810, 820, 850,870, 870, 810, 740, 810, 940, 950, 800, 810, 870]
]
```

```python
boxplot = Boxplot(size)
```

## 词云图 WordCloud

词云图，也叫文字云，是对文本中出现频率较高的“关键词”予以视觉化的展现，词云图过滤掉大量的低频低质的文本信息，使得浏览者只要一眼扫过文本就可领略文本的主旨。

```python
from pyecharts.charts import WordCloud
```

```python
wordcloud = WordCloud(size)
```

```python
data = [('足球', 22),("篮球" , 18) , ('羽毛球', 10),('乒乓球',16)]
```

```python
# 系列数据项，[(word1, count1), (word2, count2)]
wordcloud.add('词云图测试', data)
wordcloud.render_notebook()
```

### 读取数据绘图

```python
beijing = pd.read_csv('./data/beijing.csv',index_col = 0)
```

```python
beijing
```

```python
# 处理数据结果
cat_result = beijing.岗位技能.str.cat(sep = ' ')
```

```python
word_list = cat_result.split(' ')
```

```python
word_list
```

```python
# 使用集合去重
set(word_list)
```

```python
word_list_count = [] # 追加元组  ("单词名称", 次数)
for word in set(word_list):
    word_list_count.append( (word, word_list.count(word)))
```

```python
word_list_count
```

```python
wordcloud = WordCloud(opts.InitOpts(width = '800px', height='600px'))
```

```python
wordcloud.add('岗位技能词云图' , word_list_count)
wordcloud.render_notebook()
```

### 把数据分析这个单词删除掉

```python
word_list_set = set(word_list)
```

```python
word_list_set.remove('数据分析')
```

```python
word_list_count = [] # 追加元组  ("单词名称", 次数)
for word in word_list_set:
    word_list_count.append( (word, word_list.count(word)))
```

```python
wordcloud = WordCloud(opts.InitOpts(width = '800px', height='600px'))
```

```python
wordcloud.add('岗位技能词云图' , word_list_count)
wordcloud.render_notebook()
```

### 对词云图做一个调整

```python
wordcloud = WordCloud(opts.InitOpts(width = '800px', height='600px'))
```

```python
# 词云图轮廓，有 'circle', 'cardioid', 'diamond', 'triangle-forward', 'triangle', 'pentagon', 'star' 可选
```

```python
wordcloud.add('岗位技能词云图' , word_list_count ,rotate_step = 150 )
wordcloud.set_global_opts(title_opts=opts.TitleOpts('词云图绘制'))
wordcloud.render_notebook()
```
