## Line（折线/面积图）

折线图是用折线将各个数据点标志连接起来的图表，用于展现数据的变化趋势。

### 导包

```python
import pyecharts.options as opts
from pyecharts.faker import  Faker
from pyecharts.charts import Line
```

### 绘制一个折线图

```python
attr = ["第1天", "第2天", "第3天", "第4天", "第5天", "第6天"]
v1 = [5, 20, 36, 10, 10, 100]
v2 = [55, 60, 16, 20, 15, 80]
```

```python
line = Line()
```

```python
line.add_xaxis(attr)
line.add_yaxis("一号门店", v1)
line.render_notebook()
```

### 画两条折线图

```python
line = Line()
line.add_xaxis(attr)
line.add_yaxis("一号门店", v1)
line.add_yaxis("二号门店", v2)
line.render_notebook()
```

### 添加数据点和数据线

```python
line = Line()
```

```python
line.add_xaxis(attr)
line.add_yaxis("一号门店", v1 , 
    markpoint_opts= opts.MarkPointOpts(
            data = [opts.MarkPointItem(type_='max', name = '最大值')]))
line.add_yaxis("二号门店", v2,
              markline_opts= opts.MarkLineOpts(
              data = [opts.MarkLineItem(type_ ="average" , name = '平均值')]
              )
              )
line.set_series_opts()
line.render_notebook()
```

### 绘制平滑曲线

```python
line = Line()
line.add_xaxis(attr)
line.add_yaxis("一号门店", v1 , is_smooth=True,
    markpoint_opts= opts.MarkPointOpts(
            data = [opts.MarkPointItem(type_='max', name = '最大值')]))
line.add_yaxis("二号门店", v2, is_smooth=True,
              markline_opts= opts.MarkLineOpts(
              data = [opts.MarkLineItem(type_ ="average" , name = '平均值')]
              )
              )
line.set_series_opts()
line.render_notebook()
```

### 更改标记点形状

```python
line = Line()
line.add_xaxis(attr)
line.add_yaxis("一号门店", v1 ,is_smooth=True,
    markpoint_opts= opts.MarkPointOpts(symbol = 'diamond',
            data = [opts.MarkPointItem(type_='max', name = '最大值')]))
line.add_yaxis("二号门店", v2, is_smooth=True,
              markline_opts= opts.MarkLineOpts(
              data = [opts.MarkLineItem(type_ ="average" , name = '平均值')]
              ),markpoint_opts = opts.MarkPointOpts(symbol = 'circle',
            data = [opts.MarkPointItem(type_='min', name = '最小值')])
            
              )
line.set_series_opts()
line.render_notebook()
```

### 阶梯图

```python
line = Line()
line.add_xaxis(attr)
line.add_yaxis("一号门店", v1 ,is_step= True,
    markpoint_opts= opts.MarkPointOpts(symbol = 'diamond',
            data = [opts.MarkPointItem(type_='max', name = '最大值')]))
# line.add_yaxis("二号门店", v2,is_step= True,
#               markline_opts= opts.MarkLineOpts(
#               data = [opts.MarkLineItem(type_ ="average" , name = '平均值')]
#               ),markpoint_opts = opts.MarkPointOpts(symbol = 'circle',
#             data = [opts.MarkPointItem(type_='min', name = '最小值')])
            
#               )
line.set_series_opts()
line.render_notebook()
```

## Kline/Candlestick（K线图）

* 红涨蓝跌
* 日K线图又称阴阳烛，最初是日本米商用来表示米价涨跌状况的工具，后来引入股市，并逐渐风行于东南亚地区。K线图以其直观、立体感强的特点而深受投资者欢迎。
* 股市及期货市场中的K线图的画法包含四个数据，即开盘价、最高价、最低价、收盘价，所有的k线都是围绕这四个数据展开，反映大势的状况和价格信息。如果把每日的K线图放在一张纸上，就能得到日K线图，同样也可画出周K线图、月K线图。
