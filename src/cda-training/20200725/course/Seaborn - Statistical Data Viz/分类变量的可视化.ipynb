## 绘制多层面板分类图

### FacetGrid
当你想要在数据集的子集中单独地可视化变量的分布或多个变量之间的关系时，FacetGrid类是有用的。

```python
sns.set(style="darkgrid")
```

```python
g = sns.FacetGrid(tips, col="time")
# g
g.map(plt.hist, "tip");
```

```python
sns.FacetGrid?
# g.map?
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker", size=5)
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend();
```

```python
g = sns.FacetGrid(tips, row="sex", col="time",
                  margin_titles=True
                 )
bins = np.linspace(0, 60, 15)
g.map(plt.hist, "total_bill", color="steelblue", bins=bins, lw=0);
```

```python
ordered_days = tips.day.value_counts().index
g = sns.FacetGrid(tips, row="day",
                  row_order=ordered_days,  # 指定行的顺序
                  size=2, aspect=4)
g.map(sns.distplot, "total_bill", hist=False, rug=True);
```

```python
g = sns.FacetGrid(tips, col="day", size=4, aspect=.9)
g.map(sns.barplot, "sex", "total_bill",
      order=["Female", "Male"]  # 指定横坐标顺序
     );
```

```python
g = sns.FacetGrid(tips, col="day", hue="smoker")
g.map(sns.distplot, "total_bill")
g.add_legend();
```

### factorplot方法结合分类变量可视化方法及FacetGrid类更加方便，快捷地实现多层面板分类图

```python
sns.factorplot(x="day", y="total_bill", hue="smoker", data=tips);  # 默认是点图
```

```python
# sns.factorplot?
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker", data=tips, kind="bar");
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker",
               col="time", data=tips, kind="swarm");
```

```python
sns.factorplot(x="time", y="total_bill", hue="smoker",
               col="day", data=tips, kind="box",
               size=4, aspect=.7);
```
