# Grid 类：并行显示多张图

用户可以自定义结合 Line/Bar/Kline/Scatter/EffectScatter/Pie/HeatMap/Boxplot 图表，将不同类型图表画在多张图上。第一个图需为 有 x/y 轴的图，即不能为 Pie，其他位置顺序任意。

```python
from pyecharts import Line, Pie, Grid  # 引入 Grid 类

line = Line("折线图示例", width=1200)
attr = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
line.add("最高气温", attr, [11, 11, 15, 13, 12, 13, 10],
         mark_point=["max", "min"], mark_line=["average"])
line.add("最低气温", attr, [1, -2, 2, 5, 3, 2, 0],
         mark_point=["max", "min"], mark_line=["average"],
         legend_pos="20%")

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [11, 12, 13, 10, 10, 10]
pie = Pie("饼图示例", title_pos="55%")
pie.add("", attr, v1, radius=[35, 65], center=[65, 50],
        legend_pos="80%", legend_orient='vertical')

grid = Grid(width=1000)
grid.add(line, grid_right="55%")
grid.add(pie, grid_left="70%")
# grid.render()
grid
```

```python
from pyecharts import Line, Pie, Grid  # 引入 Grid 类

line = Line("折线图示例", width=1200)
attr = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
line.add("最高气温", attr, [11, 11, 15, 13, 12, 13, 10],
         mark_point=["max", "min"], mark_line=["average"])
line.add("最低气温", attr, [1, -2, 2, 5, 3, 2, 0],
         mark_point=["max", "min"], mark_line=["average"],
         legend_pos="20%")

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [11, 12, 13, 10, 10, 10]
pie = Pie("饼图示例", title_pos="55%")
pie.add("", attr, v1, radius=[35, 65], center=[65, 50],
        legend_pos="80%", legend_orient='vertical')

grid = Grid(width=1000)
grid.use_theme('dark')  # Grid 里面更换主题，单个图表的主题会被Grid类实例对象的主题覆盖
grid.add(line, grid_right="55%")
grid.add(pie, grid_left="70%")
# grid.render()
grid
```

# Overlap 类：结合不同类型图表叠加画在同张图上

用户可以自定义结合 Line/Bar/Kline, Scatter/EffectScatter 图表，将不同类型图表画在一张图上。利用第一个图表为基础，往后的数据都将会画在第一个图表上。

```python
from pyecharts import Line, Bar, Overlap  # 引入 Overlap 类

attr = ["{}月".format(i) for i in range(1, 13)]
v1 = [2.0, 4.9, 7.0, 23.2, 25.6, 76.7, 135.6, 162.2, 32.6, 20.0, 6.4, 3.3]
v2 = [2.6, 5.9, 9.0, 26.4, 28.7, 70.7, 175.6, 182.2, 48.7, 18.8, 6.0, 2.3]
v3 = [2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 20.3, 23.4, 23.0, 16.5, 12.0, 6.2]

bar = Bar(width=1200, height=600)
bar.add("蒸发量", attr, v1)
bar.add("降水量", attr, v2, yaxis_formatter=" ml",
        yaxis_interval=50, yaxis_max=250)

line = Line()
line.add("平均温度", attr, v3, yaxis_formatter=" °C", yaxis_interval=5)

overlap = Overlap()  # 实例化 Overlap 类
# 默认不新增 x y 轴，并且 x y 轴的索引都为 0
# overlap.use_theme('vintage')  # 自定义overlap主题，注意这里某些主题会有bug，如macarons

overlap.add(bar)
# 新增一个 y 轴，此时 y 轴的数量为 2，第二个 y 轴的索引为 1（索引从 0 开始），所以设置 yaxis_index = 1
# 由于使用的是同一个 x 轴，所以 x 轴部分不用做出改变
# overlap.add(line)
overlap.add(line, yaxis_index=1, is_add_yaxis=True)
# overlap.render()
overlap
```

# Timeline 类：提供时间线轮播多张图

```python
import pandas as pd
from random import randint, seed
seed(1026)
from pyecharts import Bar, Timeline  # 引入Bar类， Timeline类

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋"]
timeline = Timeline(is_auto_play=True, timeline_bottom=0, width=1000)
# timeline.use_theme('macarons')  # 自定义timeline主题

for y in pd.date_range('01/01/2010', '05/01/2018', freq='Y').to_period("Y"):
    bar = Bar("{} 年销量".format(y), "数据纯属虚构")
    for q in ["春季","夏季","秋季","冬季"]:
        bar.add(q, attr, [randint(100, 1000) for _ in range(5)], is_legend_show=True,
#                 is_stack=True,
                bar_category_gap='30%',
                is_toolbox_show=True,  # 被覆盖，不再work
                is_more_utils=True,  # 被覆盖，不再work
               )
    timeline.add(bar, '{} 年'.format(y))  #  接受两个参数，第一个为图实例，第二个为时间线的“时间点”

timeline
```

```python
import pandas as pd
from random import randint, seed
seed(1026)
from pyecharts import Pie, Timeline

attr = ["女装", "男装", "女鞋", "美妆", "手机", "首饰"]

timeline = Timeline(
    is_auto_play=True,
#                     is_loop_play=False,  # 是否循环播放，默认为 True
    is_loop_play=True,  # 是否循环播放，默认为 True
                    is_rewind_play=True,  # 是否倒回播放，默认为 Flase
                    is_timeline_show=True,
#     is_timeline_show=False,
                    timeline_play_interval=300,  # 播放的速度（跳动的间隔），单位毫秒（ms）。
                    timeline_bottom=0,
                    width=1000)

for y in pd.date_range('01/01/2010', '05/01/2018', freq='Q').to_period("Q"):
    pie = Pie("{} 销量比例".format(y), "数据纯属虚构")
    pie.add("", attr, [randint(10, 100) for _ in range(6)],
            is_label_show=True, radius=[30, 55], rosetype='radius')
    timeline.add(pie, '{} 年'.format(y))

# timeline.render()
timeline
```

## Timeline & Overlap

```python
from pyecharts import Bar, Line, Timeline, Overlap
from random import randint, seed
seed(1026)

attr = ["女装", "男装", "女鞋", "美妆", "手机", "首饰"]

m = 12
timeline = Timeline(is_auto_play=True,timeline_bottom=0)
for i in ["{}月".format(i) for i in range(1, m+1)]:
    bar = Bar("{}份数据".format(i), "数据纯属虚构")
    bar.add("bar", attr, [randint(100, 500) for _ in range(6)])
    line = Line()
    line.add("line", attr, [randint(500, 800) for _ in range(6)])
    overlap = Overlap()
    overlap.add(bar)
    overlap.add(line)
    timeline.add(overlap, i)
timeline
```
