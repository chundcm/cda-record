# Pyecharts配置项的使用

在不同的图形中,配置的用法几乎都差不多,所以我们单独的拿出一节课来学习如何添加各种配置.



```python
# 用柱状图作为例子
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
cloths = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
name = '商家A'
data = [5, 20, 36, 10, 75, 90]
```

## 全局配置项

### 初始化配置项

#### 改变图片宽和高

```python
# 类的实例化,实例化一个柱状图的对象
bar = Bar(init_opts = opts.InitOpts(width = '450px',
    height = '250px'))

# 添加横轴和纵轴的数据
bar.add_xaxis(cloths)
bar.add_yaxis(name, data)
# 在jupyter 中显示
bar.render_notebook()
```

#### 修改网页标题

```python
# 类的实例化,实例化一个柱状图的对象
bar = Bar(init_opts = opts.InitOpts(width = '450px',
    height = '250px', page_title = '全局配置测试'))

# 添加横轴和纵轴的数据
bar.add_xaxis(cloths)
bar.add_yaxis(name, data)
# 在jupyter 中显示
bar.render("bar测试.html")
```

#### 修改图表主题

```python
# 内置主题类型可查看 pyecharts.globals.ThemeType
from pyecharts.globals import ThemeType
```

```python
# 查看有哪些内置主题
dir(ThemeType)[:16]
```

```python
# 类的实例化,实例化一个柱状图的对象
bar = Bar(init_opts = opts.InitOpts(width = '450px',
    height = '250px', theme = ThemeType.CHALK))

# 添加横轴和纵轴的数据
bar.add_xaxis(cloths)
bar.add_yaxis(name, data)
# 在jupyter 中显示
bar.render_notebook()
```

```python
# 类的实例化,实例化一个柱状图的对象
bar = Bar(init_opts = opts.InitOpts(width = '450px',
    height = '250px', theme = ThemeType.DARK))

# 添加横轴和纵轴的数据
bar.add_xaxis(cloths)
bar.add_yaxis(name, data)
# 在jupyter 中显示
bar.render_notebook()
```

#### 修改背景颜色

```python
# 类的实例化,实例化一个柱状图的对象
bar = Bar(init_opts = opts.InitOpts(width = '450px',
    height = '250px', bg_color = 'yellow'))

# 添加横轴和纵轴的数据
bar.add_xaxis(cloths)
bar.add_yaxis(name, data)
# 在jupyter 中显示
bar.render_notebook()
```
