# KMeans

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs   # 随机生成的混杂数据

#自己创建数据集
X, y = make_blobs(n_samples=500,n_features=2,centers=4,random_state=1)  # 500个散点，特征数量为2个，4个类别
y
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1]
            ,marker='o' #点的形状
            ,s=8 #点的大小
           );
```

```python
#如果我们想要看见这个点的分布，怎么办？
color = ["red","pink","orange","gray"]
for i in range(4):
    plt.scatter(X[y==i, 0], X[y==i, 1]
            ,marker='o' #点的形状
            ,s=8 #点的大小
            ,c=color[i]
               )
```

```python
from sklearn.cluster import KMeans

n_clusters = 3   # 设定初始质心的个数，簇的个数

cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)    

#重要属性labels_，查看聚好的类别，每个样本所对应的类
y_pred = cluster.labels_
y_pred
```

```python
#KMeans因为并不需要建立模型或者预测结果，因此我们只需要fit就能够得到聚类结果了
#KMeans也有接口predict和fit_predict，表示学习数据X并对X的类进行预测
#但所得到的结果和我们不调用predict，直接fit之后调用属性labels一模一样
pre = cluster.fit_predict(X)
pre == y_pred
```

```python
#我们什么时候需要predict呢?当数据量太大的时候！
#其实我们不必使用所有的数据来寻找质心，少量的数据就可以帮助我们确定质心了
#当我们数据量非常大的时候，我们可以使用部分数据来帮助我们确认质心
#剩下的数据的聚类结果，使用predict来调用

cluster_smallsub = KMeans(n_clusters=n_clusters, random_state=0).fit(X[:200])
y_pred_ = cluster_smallsub.predict(X)
y_pred == y_pred_
#但这样的结果，肯定与直接fit全部数据会不一致。有时候，当我们不要求那么精确，或者我们的数据量实在太大，那我们可以使用这样的方法。
```

```python
#重要属性cluster_centers_，查看质心
centroid = cluster.cluster_centers_
centroid
```

```python
#重要属性inertia_，查看总距离平方和
inertia = cluster.inertia_
inertia
```

## 轮廓系数

```python
from sklearn.metrics import silhouette_score
```

```python
score=[]
for i in range(2,10):
    km = KMeans(n_clusters=i).fit(X)
    score.append(silhouette_score(X,km.labels_))
```

```python
score
```

```python
plt.plot(range(2,10),score,marker='o')     # 找的是次高点
```

# DBSCAN

```python
from sklearn.datasets import make_moons  # 创建月亮型的数据
X,y = make_moons(200, noise = 0.05, random_state=0)
```

```python
X.shape
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
# 用K-Means
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
kmeans.predict(X)
labels
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.scatter(X[:,0],X[:,1], c = labels)  
plt.plot(centers[:,0], centers[:,1],'ro')

# 很明显K-Means并没有准确的捕捉数据集背后客观分类规律，接下来尝试用DBSCAN进行聚类
```

```python
# 在调用sklearn中DBSCAN算法时，需要设置eps半径长度以及核心点的判别条件
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.2, min_samples=2)
db.fit(X)
```

```python
db.labels_
```

```python
plt.scatter(X[:,0],X[:,1], c = db.labels_)
```

## 非凸稠密数据

```python
from sklearn.datasets import make_circles
import numpy as np
```

```python
x1, y1 = make_circles(n_samples=2000, factor=0.5, noise=.05)
x2, y2 = make_blobs(n_samples=1000, centers=[[1.2,1.2]], cluster_std=[[.1]])

x = np.concatenate((x1, x2))
plt.scatter(x[:, 0], x[:, 1], marker='o')
plt.show()
```

```python
# K-Means
y_pred = KMeans(n_clusters=3).fit_predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.show()
```

```python
# DBSCAN
y_pred = DBSCAN(eps = 0.2,min_samples=10).fit_predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.show()
```

```python
# Anisotropicly distributed data
random_state = 170
X, y = make_blobs(n_samples=1500, random_state=random_state)#各向异性点状数据
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)

plt.scatter(X_aniso[:,0], X_aniso[:, 1], marker='o')
plt.show()
```

```python
# K-Means
y_pred = KMeans(n_clusters=3).fit_predict(X_aniso)
plt.scatter(X_aniso[:,0], X_aniso[:, 1], c=y_pred)
plt.show()
```

```python
# DBSCAN
y_pred = DBSCAN(eps = 0.5,min_samples=30).fit_predict(X_aniso)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.show()
```

# KMeans算法总结

- kmeans聚类算法属于无监督学习，没有标签
- kmeans原理：先随机选择K个质心，根据样本到质心的距离将样本分配到最近的簇中，然后根据簇中的样本更新质心，再次计算距离重新分配簇，直到质心不再发生变化，迭代结束
- kmeans的Python实现
- kmeans的sklearn实现

> from sklearn.cluster import KMeans  #导入kmeans函数   
>> 重要参数：    
>> n-clusters #簇的个数   
>> init #初始化质心的方法，默认是Kmeans++   
>> random_state #控制每次质心随机初始化的随机数种子   
>> n_init  #使用不同的质心随机初始化的种子来运行k-means算法的次数
>> max_iter 最大迭代次数    
>> tol  #容差，两次迭代间Inertia下降的量    

> cluster = KMeans(n_clusters=k).fit(X) #实例化    

> 查看模型的各种属性   
>> cluster.labels_ #查看聚好的类别，每个样本所对应的类   
>> cluster.cluster_centers_ #查看质心   
>> cluster.inertia_  #查看总距离平方和   

- 聚类算法的评估指标：轮廓系数
