```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs   # 随机生成的混杂数据

#自己创建数据集
X, y = make_blobs(n_samples=500,n_features=2,centers=4,random_state=1)  # 500个散点，特征数量为2个，4个类别
y
```

```python
plt.scatter(X[:,0],X[:,1])  # 没有标签
```

```python
for i in range(4):
    plt.scatter(X[y==i,0],X[y==i,1])    # 有标签
```

### sklearn-Kmeans

```python
from sklearn.cluster import KMeans
```

```python
# 实例化
model = KMeans(n_clusters=6)
# 训练数据
model = model.fit(X)
# 输出聚类标签结果，每一个样本所对应的类
model.labels_
```

```python
# 查看质心
model.cluster_centers_
```

```python
# 评估指标：总距离平方和（总sse  总误差平方和）
model.inertia_
```

```python
ypred = model.labels_
```

```python
pd.Series(ypred).unique()
```

```python
# 可视化看看~
for i in range(6):
    plt.scatter(X[ypred==i,0],X[ypred==i,1])
```

```python
for i in range(4):
    plt.scatter(X[y==i,0],X[y==i,1])    # 有标签
```

### 怎么去寻找最优的簇数？

#### 方法一：基于SSE的手肘法

```python
sse = []   # 初始化列表，用来存储不同簇数下sse的值
for i in range(2,11):
    model = KMeans(n_clusters=i,random_state=2).fit(X)
    sse.append(model.inertia_)
```

```python
sse
```

```python
plt.plot(range(2,11),sse,marker='o')
```

#### 方法二：轮廓系数

```python
from sklearn.metrics import silhouette_score
```

```python
score = []
for i in range(2,11):
    model = KMeans(n_clusters=i).fit(X)
    score.append(silhouette_score(X,model.labels_))
```

```python
plt.plot(range(2,11),score,marker='o')   # 次高点
```

```python
model = KMeans(n_clusters=4,init='random',max_iter=10).fit(X)
```
